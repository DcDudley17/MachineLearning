```python
import numpy as np
import matplotlib.pyplot as plt
import csv

%matplotlib inline
```

```python
import numpy as np

def add_noise(func, x,noise_level):
    y = func(x)
    noise = np.random.normal(0, noise_level, size=y.shape)
    return y + noise
```

```python
def func(x):
    return (np.log(x)+5*np.sin(x))
xs = np.linspace(0.01,10,1001)

output = np.zeros(shape=(1001,3))

output[:, 0] = np.arange(1, 1002)       # Indices 1 to 1001
output[:, 1] = xs                       # x values
output[:, 2] = add_noise(func,xs,1)     # y = func(x) values

plt.plot(xs,add_noise(func,xs,1),'r.')
plt.show()

# Define the output CSV file name
csv_file = "func1_output.csv"

# Open the CSV file in write mode
with open(csv_file, "w", newline="") as file:
    # Create a CSV writer object
    writer = csv.writer(file)

    # Write the header row (optional, but good practice)
    writer.writerow(["Index", "X", "Y"])

    # Iterate over the rows in the 'output' array and write to the CSV file
    for row in output:
        writer.writerow(row)

print(f"Data successfully written to {csv_file}")
```
